```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file
df = pd.read_csv("admit_rates.csv")

# Plot
plt.figure(figsize=(10, 6))

plt.plot(df["Year"], df["Admit Rate Computer Science(%)"], linewidth=2, label="Computer Science")

plt.title("Admissions Trends for CS at UC Berkeley (20122023)", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Admit Rate (%)", fontsize=12)
plt.xticks(df["Year"], rotation=45)
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()
```
